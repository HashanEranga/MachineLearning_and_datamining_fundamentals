## Stemming
Stemming is the process of reducing a word to its word stem also known as stem

```python
word_list = [
    "Flying", "Flew", "Flier",
    "Sided", "Side", "Siderial",
    "Messed", "Messing", "Matcher"
]
```

### PorterStemmer

```python
from nltk.stem import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
for word in word_list:
    print(word, "->", ps.stem(word))
```

### RegexpStemmer

```python
from nltk.stem import RegexpStemmer
```

```python
# Remove common suffixes using regex
regex_stemmer = RegexpStemmer('ing$|s$|ed$|e$', min=4)
```

```python
for word in word_list:
    print(word, "->", regex_stemmer.stem(word))
```

### SnowballStemmer

```python
from nltk.stem import SnowballStemmer
```

```python
snowball_stemmer = SnowballStemmer('english')
```

```python
for word in word_list:
    print(word, "->", snowball_stemmer.stem(word))
```

### LancasterStemmer
Lancaster stemmer is more aggressive than Porter stemmer, producing shorter stems. It uses an iterative algorithm with over 120 rules.

```python
from nltk.stem import LancasterStemmer
```

```python
lancaster_stemmer = LancasterStemmer()
```

```python
for word in word_list:
    print(word, "->", lancaster_stemmer.stem(word))
```
