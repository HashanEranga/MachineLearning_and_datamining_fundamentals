### BOW - Bag Of Words

```python
import re

import pandas as pd
from pyexpat.errors import messages
```

```python
df = pd.read_csv('spam.csv', encoding='latin1', names=['label', 'message', 'test', 'test1', 'test2'])
df.head()
```

```python
df = df.iloc[1:, :-3]
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
import nltk
import re
from nltk.stem.porter import PorterStemmer
from nltk.corpus import stopwords
ps = PorterStemmer()
nltk.download('stopwords')
stop_words = stopwords.words('english')
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=2500, binary=False, ngram_range=(2,3))
```

```python
# data cleaning and preprocessing
corpus = []

for message in df['message']:
    review = re.sub('[^a-zA-Z]', ' ',message)
    review = review.lower()
    review = ' '.join([ps.stem(word) for word in review.split() if word not in stop_words])
    corpus.append(review)
```

```python
corpus
```

```python
corpus_array = cv.fit_transform(corpus).toarray()
```

```python
corpus_array
```

### BOW - using lemmer

```python
# import
from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet
lemmer = WordNetLemmatizer()
```

```python
df = pd.read_csv('spam.csv', encoding='latin1', names=['label', 'message', 'test', 'test1', 'test2'])
df = df.iloc[1:, :-3]
df.head()
```

```python
def get_wordnet_pos(word):
    """Map POS tag to the first character lemmatize () accepts"""
    tag = nltk.pos_tag([word])[0][1][0].upper()
    tag_dict = {"J": wordnet.ADJ,
                "N": wordnet.NOUN,
                "V": wordnet.VERB,
                "R": wordnet.ADV}
    return tag_dict.get(tag, wordnet.NOUN)
```

```python
# data cleaning and preprocessing
corpus = []

for message in df['message']:
    review = re.sub('[^a-zA-Z]', ' ',message)
    review = review.lower()
    review = ' '.join([lemmer.lemmatize(word, get_wordnet_pos(word)) for word in review.split() if word not in stop_words])
    corpus.append(review)
```

```python
corpus
```

```python
corpus_lemmer_array = cv.fit_transform(corpus).toarray()
```
