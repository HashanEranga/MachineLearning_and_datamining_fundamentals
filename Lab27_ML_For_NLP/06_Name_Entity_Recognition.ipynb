### Name Entity Recognition

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.tag import pos_tag
from nltk.chunk import ne_chunk
nltk.download('maxent_ne_chunker_tab')
nltk.download('words')
```

```python
sentense = "Apple Inc., headquartered in Cupertino, California, reported quarterly earnings on July 28th."
```

```python
tokenizer = word_tokenize(sentense.lower())
tokenizer
```

```python
pos_tags = pos_tag(tokenizer)
pos_tags
```

```python
ner_chunked = ne_chunk(pos_tags)
ner_chunked.draw()
```
