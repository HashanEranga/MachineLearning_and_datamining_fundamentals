## Finding Outliers

```python
import numpy as np
import seaborn as sns
```

```python
values_list = [23,34,45,56,34,67,67,89,45,23,67,78,34,67,78,89,45,34,23,56,78,89]

minimum, Q1, median, Q3, maximum = np.quantile(values_list, [0, 0.25, 0.5, 0.75, 1])
IQR = Q3 - Q1
```

```python
print(f'Minimum value {minimum}')
print(f'Q1 value {Q1}')
print(f'Q2 value {median}')
print(f'Q3 value {Q3}')
print(f'Maximum value {maximum}')
print(f'Interquartile range {IQR}')
```

## 

```python
lower_floor = Q1 - 1.5 * IQR
upper_floor = Q3 + 1.5 * IQR

print(f'Lower floor {lower_floor}')
print(f'Upper floor {upper_floor}')
```

## Plotting on a boxplot

```python
sns.boxplot(data=values_list)
```

## plotting a boxplot with outliers

```python
value_list_with_outliers = values_list + [140, 190, 140, 160, 180, -20, -140]

sns.boxplot(data=value_list_with_outliers)
```
